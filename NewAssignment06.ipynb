## Estimation And Confidence Intervals

```python
import numpy as np
from scipy import stats

data = np.array([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29])

n = len(data)
x_bar = np.mean(data)
s = np.std(data, ddof=1)
se = s / np.sqrt(n)

alpha = 0.01
df = n - 1
t_critical = stats.t.ppf(1 - alpha/2, df)
margin_t = t_critical * se
ci_t_lower = x_bar - margin_t
ci_t_upper = x_bar + margin_t

print("===== Part A: 99% CI Using Sample Standard Deviation (t-distribution) =====")
print(f"Sample Size (n)          : {n}")
print(f"Sample Mean (x̄)          : {x_bar:.4f}")
print(f"Sample Std Dev (s)       : {s:.4f}")
print(f"Standard Error (SE)      : {se:.4f}")
print(f"Degrees of Freedom (df)  : {df}")
print(f"t-Critical Value         : {t_critical:.4f}")
print(f"Margin of Error          : {margin_t:.4f}")
print(f"99% Confidence Interval  : ({ci_t_lower:.4f}, {ci_t_upper:.4f})")
```

```python
sigma = 0.2
se_z = sigma / np.sqrt(n)
z_critical = stats.norm.ppf(1 - alpha/2)
margin_z = z_critical * se_z
ci_z_lower = x_bar - margin_z
ci_z_upper = x_bar + margin_z

print("===== Part B: 99% CI Using Known Population Standard Deviation (z-distribution) =====")
print(f"Population Std Dev (σ)   : {sigma}")
print(f"Standard Error (SE)      : {se_z:.4f}")
print(f"z-Critical Value         : {z_critical:.4f}")
print(f"Margin of Error          : {margin_z:.4f}")
print(f"99% Confidence Interval  : ({ci_z_lower:.4f}, {ci_z_upper:.4f})")
```
