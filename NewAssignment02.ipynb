```python
#Q1.List Operations

my_list = [10, 20, 30, 40, 50]

my_list.append(60)
print("After append:", my_list)

my_list.extend([70, 80])
print("After extend:", my_list)

my_list.insert(2, 25)
print("After insert:", my_list)

my_list.remove(25)
print("After remove:", my_list)

my_list.pop()
print("After pop:", my_list)

print("Index of 30:", my_list.index(30))
print("Count of 10:", my_list.count(10))

my_list.sort()
print("After sort:", my_list)

my_list.reverse()
print("After reverse:", my_list)

my_list.clear()
print("After clear:", my_list)
```

```python
student_names = ("Aarav", "Priya", "Rohit")
print("Tuple:", student_names)

try:
    student_names[1] = "Sneha"
except TypeError as e:
    print("Error:", e)

# Explanation:
# Tuples are immutable in Python.
# Once created, their values cannot be changed.
# That is why we get a TypeError when we try to modify an element.
```

```python
#Q3.Set
my_set = {1, 2, 2, 3, 4, 4, 5}
print("Set:", my_set)

# Explanation:
# A set automatically removes duplicate values.
# Even though we added 2 and 4 twice, the set stores only unique values.
```

```python
#Q7.Positive Negative
number = 5

if number > 0:
    if number % 2 == 0:
        print(number, "is Positive and Even")
    else:
        print(number, "is Positive but Odd")
elif number == 0:
    print("The number is Zero")
else:
    print("The number is Negative")
```

```python
#Q11.DataFrame
import pandas as pd

data = {
    'Name':   ['Dipak', 'Aakash', 'Mayur', 'Chetan', 'Rohit',
                'Sakshi', 'Nikita', 'Neha', 'Prachi', 'Jagruti'],
    'Age':    [23, 25, 22, 30, 28, 35, 27, 24, 29, 31],
    'City':   ['Mumbai', 'Nashik', 'Pune', 'Jalgaon', 'Amravati',
                'Kolhapur', 'Delhi', 'Nagpur', 'Chennai', 'Kolkata'],
    'Salary': [30000, 45000, None, 50000, 38000,
                60000, None, 42000, 35000, 55000]
}

df = pd.DataFrame(data)

print("--- .info() ---")
print(df.info())

print("\n--- .describe() ---")
print(df.describe())

print("\n--- Name and City columns ---")
print(df[['Name', 'City']])

df = df.drop(columns=['City'])
print("\n--- After dropping City ---")
print(df)

df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
print("\n--- After filling missing Salary with mean ---")
print(df)

df = df.drop_duplicates()
print("\n--- After removing duplicates ---")
print(df)
```
